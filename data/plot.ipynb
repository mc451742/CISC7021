```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum()
```

```python
# data for the chart
iterations = [0, 1, 2, 3]
categories = ["Que.", "Res.", "Ref.", "Sel.", "Aff.", "Pro.", "Inf.", "Oth."]

# define ∆P_t (%) data for Direct-Refine and Self-Refine
direct_refine_data = [
    [0, -30, -60, -90],  # Que.
    [0, -20, -40, -60],  # Res.
    [0, 10, 20, 30],     # Ref.
    [0, 15, 25, 35],     # Sel.
    [0, 20, 25, 30],     # Aff.
    [0, -10, -20, -30],  # Pro.
    [0, -60, -70, -80],  # Inf.
    [0, -5, -10, -15],   # Oth.
]

self_refine_data = [
    [0, 20, 30, 40],     # Que.
    [0, -10, -20, -30],  # Res.
    [0, 5, 10, 15],      # Ref.
    [0, 15, 30, 40],     # Sel.
    [0, 25, 35, 40],     # Aff.
    [0, -15, -30, -45],  # Pro.
    [0, -60, -75, -90],  # Inf.
    [0, -5, -10, -15],   # Oth.
]

# colors for each category
colors = [
    "red", "orange", "green", "yellowgreen", "cyan", "blue", "darkblue", "gray"
]

# initialize the figure and axes
fig, axes = plt.subplots(1, 2, figsize=(10, 6), sharey=True)

# plot Direct-Refine
ax1 = axes[0]
for i, data in enumerate(direct_refine_data):
    ax1.plot(iterations, data, marker='o', color=colors[i], label=categories[i])

ax1.axhline(0, color='black', linestyle='--', linewidth=0.8) # horizontal line at 0
ax1.set_title("Direct-Refine")
ax1.set_xlabel("Iteration")
ax1.set_ylabel("∆P_t (%)")
ax1.legend(title="Categories", bbox_to_anchor=(1.05, 1), loc='upper left')
ax1.grid(True)

# plot Self-Refine
ax2 = axes[1]
for i, data in enumerate(self_refine_data):
    ax2.plot(iterations, data, marker='o', color=colors[i], label=categories[i])

ax2.axhline(0, color='black', linestyle='--', linewidth=0.8) # horizontal line at 0
ax2.set_title("Self-Refine")
ax2.set_xlabel("Iteration")
ax2.grid(True)

# adjust layout
plt.tight_layout()
plt.show()
```

```python
# data for the radar charts
categories = ["Que.", "Res.", "Ref.", "Sel.", "Aff.", "Pro.", "Inf.", "Oth."]
num_vars = len(categories)

# data for each model; these are example values
gpt4_data = [20, 15, 30, 25, 20, 10, 15, 10]
chatgpt_data = [15, 10, 25, 20, 15, 5, 10, 5]
tulu_data = [10, 15, 20, 25, 15, 10, 15, 20]
llama2_data = [25, 20, 15, 20, 10, 5, 10, 15]

# colors for each radar chart
colors = ["mediumseagreen", "cornflowerblue", "orchid", "gold"]

# function to create a radar chart
def create_radar_chart(ax, data, color, title):
    # compute angle for each category
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    data += data[:1] # repeat the first value to close the circle
    angles += angles[:1]

    # plot the data
    ax.fill(angles, data, color=color, alpha=0.25)
    ax.plot(angles, data, color=color, linewidth=2)

    # add category labels
    ax.set_yticks([5, 10, 15, 20, 25, 30, 35])
    ax.set_yticklabels(["5", "10", "15", "20", "25", "30", "35"], color="gray", fontsize=8)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=10, color="dimgray")

    # set title
    ax.set_title(title, position=(0.5, 1.1), fontsize=10, color="dimgray")

# create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, subplot_kw={"projection": "polar"}, figsize=(8, 8))

# plot each radar chart
create_radar_chart(axs[0, 0], gpt4_data, colors[0], "GPT-4 (2-shot)")
create_radar_chart(axs[0, 1], chatgpt_data, colors[1], "ChatGPT (2-shot)")
create_radar_chart(axs[1, 0], tulu_data, colors[2], "Tulu (2-shot)")
create_radar_chart(axs[1, 1], llama2_data, colors[3], "LLaMA2-70B (2-shot)")

# adjust layout
plt.tight_layout()
plt.show()
```
